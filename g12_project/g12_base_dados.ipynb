```python
# Leitura do ficheiro
import pandas as pd
import sqlite3

df = pd.read_csv('g12_Universities_Alumni.csv', sep=';', encoding='latin1')

display(df.head())
```

```python
# Separar em tabelas

universidades = df[['university_id', 'name', 'foundation_date']].drop_duplicates()
associacoes = df[['association_id', 'designation', 'objective']].drop_duplicates()
inscricoes = df[['university_id', 'association_id', 'registration_date', 'fee']]
alumni = df[['graduate_id', 'university_id', 'obs']].dropna(subset=['graduate_id'])
```

```python
# Criar ligação com a base de dados (cria ficheiro .db)
conn = sqlite3.connect('universidades_alumni.db')

# Enviar os dados para o SQLite
universidades.to_sql('Universidade', conn, if_exists='replace', index=False)
associacoes.to_sql('Associacao', conn, if_exists='replace', index=False)
inscricoes.to_sql('Inscricao', conn, if_exists='replace', index=False)
alumni.to_sql('Alumni', conn, if_exists='replace', index=False)

# Fechar a ligação
conn.close()
print("Conversão concluída com sucesso!")
```
